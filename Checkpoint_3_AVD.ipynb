###Melihat Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Melakukan Imputasi

```python
columns_to_impute = ["year", "price", "mileage", "tax", "mpg", "engineSize"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Pengecekan Ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

###Konstruksi Data

Disini saya menambahkan kolom 'car_age', 'depreciation_per_year', dan 'price_per_mile'.

```python
from datetime import datetime

df = df.copy()
current_year = datetime.now().year

# Menambahkan kolom 'car_age' (usia mobil)
df["car_age"] = current_year - df["year"]

# Menambahkan kolom 'depreciation_per_year' (tingkat penyusutan per tahun)
df["depreciation_per_year"] = df["price"] / df["car_age"].replace(0, 1)  # Hindari pembagian dengan nol

# Menambahkan kolom 'price_per_mile' (harga per jarak tempuh)
df["price_per_mile"] = df["price"] / df["mileage"].replace(0, 1)  # Hindari pembagian dengan nol

df.head()
```

###Analisis Korelasi

```python
print("Menentukan Korelasi Antar Kolom")
print(df[['car_age', 'mileage', 'depreciation_per_year', 'price']].corr())
```

Dari hasil analisis korelasi ini, bisa disimpulkan bahwa usia mobil adalah faktor utama yang mempengaruhi harga jual kembali, sementara jarak tempuh juga berpengaruh, tetapi tidak sebesar usia.

##Comparison

```python
plt.figure(figsize=(8, 5))
avg_price_by_age = df.groupby('car_age')['price'].mean()
plt.bar(avg_price_by_age.index, avg_price_by_age.values, color='blue')
plt.title('Rata-rata Harga Mobil Berdasarkan Usia')
plt.xlabel('Usia Mobil (Tahun)')
plt.ylabel('Rata-rata Harga (USD)')
plt.xticks(rotation=45)
plt.show()
```

Bar chart di atas menunjukkan perbandingan rata-rata harga mobil berdasarkan usia mobil. Sumbu x mewakili usia mobil dalam tahun, sedangkan sumbu y menunjukkan rata-rata harga mobil dalam satuan mata uang yang digunakan dalam dataset. Dari grafik, terlihat bahwa semakin bertambah usia mobil, harga rata-ratanya cenderung menurun. Mobil yang lebih baru memiliki harga yang lebih tinggi, sedangkan mobil yang lebih tua mengalami depresiasi nilai yang signifikan. Meskipun tren umum menunjukkan penurunan harga, ada beberapa usia mobil yang memiliki harga rata-rata lebih tinggi dari yang diharapkan. Hal ini bisa disebabkan oleh faktor lain seperti kondisi mobil, jumlah pemakaian (mileage), jenis transmisi, atau model mobil yang lebih populer.

Bagi pembeli, memahami pola ini dapat membantu dalam menentukan kapan waktu terbaik untuk membeli mobil bekas dengan harga yang lebih terjangkau. Sementara itu, bagi penjual, mengetahui tren ini memungkinkan mereka untuk menetapkan harga jual yang kompetitif dan mempertimbangkan kapan waktu terbaik untuk menjual mobil agar mendapatkan keuntungan yang optimal.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=30, kde=True, color='blue')
plt.title('Distribusi Harga Mobil Toyota', fontsize=14)
plt.xlabel('Harga Mobil', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.show()
```

Histogram di atas menampilkan sebaran harga mobil Toyota dalam dataset. Sumbu X mewakili nilai harga mobil, sedangkan sumbu Y menunjukkan frekuensi kemunculan harga tersebut dalam data. Dengan menggunakan 30 bins, grafik ini memberikan gambaran yang cukup detail mengenai bagaimana harga mobil tersebar, apakah sebagian besar berada pada kisaran tertentu atau tersebar secara merata. Garis Kernel Density Estimate (KDE) ditambahkan untuk menggambarkan pola distribusi secara halus, sehingga kita bisa mengidentifikasi apakah distribusinya normal, skewed, atau terdapat outlier. Informasi ini berguna bagi pembeli dan penjual untuk memahami tren pasar mobil bekas, misalnya apakah harga cenderung terkonsentrasi pada rentang tertentu atau ada nilai ekstrem yang perlu diperhatikan dalam penentuan harga atau strategi pembelian.
